```python
def assignIpAdd(device_id):
    base_ip="198.168.1."

    hostnum=2
    networkmaping={}

    for device_id in device_id:
        assignip= f"{base_ip}{hostnum}"
        networkmaping[device_id]= assignip

        hostnum+=1

        return networkmaping

d={501,502,503,504}
assignIpAddress= assignIpAdd(d)

for device, ip in assignIpAddress.items():
    print(f"Device ID:{device} -> Ip Address: {ip}")
```

```python
sequence=[]
def seq(a):
    for i in range(a):
        sequ=i*i+3*i
        sequence.append(sequ)
    return sequence

print(seq(10))
```

```python
def bat(hours):
    bate=300
    for i in range(0,hours):
        bate*=2
        print(f"bateria growth in {i} hours: {bate}")
    print(f"final growth is: {bate}")

bat(10)
```
